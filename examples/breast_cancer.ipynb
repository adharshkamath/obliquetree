```python
from obliquetree import Classifier
from obliquetree.utils import visualize_tree

from sklearn.tree import DecisionTreeClassifier


from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score

RANDOM_STATE = 42
```

```python
X, y = load_breast_cancer(return_X_y=True)

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=RANDOM_STATE)

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
clf = Classifier(use_oblique=True, max_depth=2, random_state=RANDOM_STATE, n_pair=2)

clf.fit(X_train, y_train)

train_pred = clf.predict(X_train)
train_mse = accuracy_score(y_train, train_pred)

test_pred = clf.predict(X_test)
test_mse = accuracy_score(y_test, test_pred)

print(train_mse, test_mse)
```

```python
clf_sk = DecisionTreeClassifier(max_depth=3, random_state=RANDOM_STATE)

clf_sk.fit(X_train, y_train)

train_pred = clf_sk.predict(X_train)
train_mse = accuracy_score(y_train, train_pred)

test_pred = clf_sk.predict(X_test)
test_mse = accuracy_score(y_test, test_pred)

print(train_mse, test_mse)
```
